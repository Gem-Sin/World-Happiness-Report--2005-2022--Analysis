## 2. Data Cleaning

```python
# Check missing values
df.isnull().sum()
```

ten of the 13 variables have missing values that need addressing.

Addressing Missing Values

There is a large amount of missing values in confidence_in_national_government column. I don't want to replace the missing values so creating a flag column.

FIRST MISSING

```python
# Create missing flag column
df["conf_gov_missing"] = df["confidence_in_national_government"].isnull().astype(int)
```

```python
# Check flag column
df[["confidence_in_national_government", "conf_gov_missing"]].head(10)
```

Addressing regional_indicator column which has 112 missing values

SECOND MISSING

```python
# Fill missing values with "Unknown"
df["regional_indicator"] = df["regional_indicator"].fillna("Unknown")
```

```python
# Check for missing values
df["regional_indicator"].isnull().sum()
```

gdp per capita has 20 missing values. As it is not so many and is numeric, I will impute the mean

THIRD MISSING

```python
# Impute GDP with the mean
df["log_gdp_per_capita"] = df["log_gdp_per_capita"].fillna(df["log_gdp_per_capita"].mean())
```

```python
# Check values
df["log_gdp_per_capita"].isnull().sum()
```

FOURTH MISSING

```python
df[df["healthy_life_expectancy_at_birth"].isnull()]
```

healthy_life_expectancy_at_birth has 54 NaNs so I will impute them with the mean

```python
# Impute life expectancy with the mean
df["healthy_life_expectancy_at_birth"] = df["healthy_life_expectancy_at_birth"].fillna(df["healthy_life_expectancy_at_birth"].mean())
```

```python
# Check impute
df["healthy_life_expectancy_at_birth"].isnull().sum()
```

FIFTH MISSING

```python
# Check missing values for social_support
df[df["social_support"].isnull()]
```

social_support has 13 missing values so will impute the mean

```python
# Impute social support with the mean
df["social_support"] = df["social_support"].fillna(df["social_support"].mean())
```

```python
# Check impute
df["social_support"].isnull().sum()
```

SIXTH MISSING

```python
# Check missing values for freedom_to_make_life_choices
df[df["freedom_to_make_life_choices"].isnull()]
```

```python
# Impute freedom with the mean
df["freedom_to_make_life_choices"] = df["freedom_to_make_life_choices"].fillna(
    df["freedom_to_make_life_choices"].mean())
```

```python
# Check impute
df["freedom_to_make_life_choices"].isnull().sum()
```

SEVENTH MISSING

Generosity has 73 missing values

```python
# Impute generosity with the mean
df["generosity"] = df["generosity"].fillna(df["generosity"].mean())
```

```python
# Check
df["generosity"].isnull().sum()
```

EIGTH MISSING

Perceptions_of_corruption has 116 misisng values

```python
# Impute perceptions of corruption with the mean
df["perceptions_of_corruption"] = df["perceptions_of_corruption"].fillna(
    df["perceptions_of_corruption"].mean())
```

```python
# Check
df["perceptions_of_corruption"].isnull().sum()
```

NINTH MISSING

Positive_affect has 19 missing values

```python
# Impute positive_affect with the mean
df["positive_affect"] = df["positive_affect"].fillna(df["positive_affect"].mean())
```

```python
# Check impute
df["positive_affect"].isnull().sum()
```

TENTH MISSING

Negative_affect has 16 missing

```python
# Impute negative affect with the mean
df["negative_affect"] = df["negative_affect"].fillna(df["negative_affect"].mean())
```

```python
# Check impute
df["negative_affect"].isnull().sum()
```
